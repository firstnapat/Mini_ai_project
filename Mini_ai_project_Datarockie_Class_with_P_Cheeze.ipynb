```python
# create pyttsx3 init
import pyttsx3

#Initial the speaker engine
engine = pyttsx3.init("sapi5")

#Set the rate of speed
engine.setProperty("rate", 100)

voices = engine.getProperty("voices")
# for voice in voices:
#     print(voice)
    
engine.setProperty("voice", voices[0].id)

engine.setProperty("volume", 1.0)

mp3_filename = "pyttsx3Result.mp3"



def speak(inputtext):
    engine.say(inputtext)
    engine.runAndWait()
    
inputtext = input("Enter the input text: ")
speak(inputtext)
```

# Apply PyPDF4 with Pyttsx3 (python text-to-speech)

```python
import PyPDF4
from gtts import gTTS
from IPython.display import Audio

language = "en"

to_read_file = "sample.pdf"
pdfReader = PyPDF4.PdfFileReader(to_read_file)

for pageNum in range(pdfReader.getNumPages()):
    pageObject = pdfReader.getPage(pageNum)
    text_to_say = pageObject.extractText()
    
    gtts_object = gTTS(text = text_to_say, lang=language, slow=False)
    
gtts_object.save("result.mp3")
mp3_file = "result.mp3"
Audio(mp3_file, autoplay=True)
```

# Face Detection with OpenCV
we use Haar Cascade Classifier Pretain for object detection

```python
!pip install opencv-python
```

```python
import cv2

imagePath = "people.jpg"

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
image = cv2.imread(imagePath)
grayscaleImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
detectedFaces = face_cascade.detectMultiScale(grayscaleImage)
# detectedFaces = face_cascade.detectMultiScale(grayscaleImage, 1.3, 6)
for x,y,w,h in detectedFaces:
    image = cv2.rectangle(image, (x,y), (x+w, y+h), (178,102,255), 2)
    cv2.imshow("Face Detection", image)
    cv2.waitKey(2000)
while True:
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
cv2.destroyAllWindows()
quit()
```
